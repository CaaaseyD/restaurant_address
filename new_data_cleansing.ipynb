```python
import re

def load_street_types(file_path):
    with open(file_path, 'r') as file:
        return [line.strip() for line in file]

def parse_address(address, street_types):
    # Regular expressions to extract different parts of the address
    unit_number_regex = r'^(?P<unit_number>[0-9]+[a-zA-Z]?)'
    street_number_regex = r'(?P<street_number>[0-9]+)'
    street_name_regex = r'(?P<street_name>[a-zA-Z ]+)'
    street_type_regex = r'(?P<street_type>[a-zA-Z]+)'
    suburb_state_postcode_regex = r'(?P<suburb>.*),\s*(?P<state>[a-zA-Z]+)\s+(?P<postcode>[0-9]+|[a-zA-Z]+)'

    # Extract the parts of the address using regular expressions
    unit_number = re.search(unit_number_regex, address)
    street_number = re.search(street_number_regex, address)
    street_name = re.search(street_name_regex, address)
    street_type_match = re.search(street_type_regex, address)
    suburb_state_postcode = re.search(suburb_state_postcode_regex, address)

    # Extract the postcode
    postcode = None
    if suburb_state_postcode:
        postcode_match = suburb_state_postcode.group('postcode')
        if postcode_match.isdigit():
            postcode = postcode_match

    # Extract the street type by matching it with the predefined list of street types
    street_type = None
    if street_type_match:
        possible_street_type = street_type_match.group('street_type').strip().lower()
        if possible_street_type in street_types:
            street_type = possible_street_type

    # Extract the state
    state = None
    if suburb_state_postcode:
        state_match = suburb_state_postcode.group('state').strip().lower()
        states_abbreviations = {
            'act': 'Australian Capital Territory',
            'nsw': 'New South Wales',
            'nt': 'Northern Territory',
            'qld': 'Queensland',
            'sa': 'South Australia',
            'tas': 'Tasmania',
            'vic': 'Victoria',
            'wa': 'Western Australia'
        }

        state = states_abbreviations.get(state_match, state_match.upper())

    # Return the extracted parts as a dictionary
    parsed_address = {
        "unit_number": unit_number.group('unit_number') if unit_number else None,
        "street_number": street_number.group('street_number') if street_number else None,
        "street_name": street_name.group('street_name').strip() if street_name else None,
        "street_type": street_type,
        "suburb": suburb_state_postcode.group('suburb').strip() if suburb_state_postcode else None,
        "state": state,
        "postcode": postcode
    }
    return parsed_address

# Load the street types from the file
street_types_file_path = 'australian-postcodes.csv'
street_types = load_street_types(street_types_file_path)

# Sample data
sample_data = [
    "7 Brrack St, Sydney, NSW 2000",
    "25 Mntague St, Sth Brisbane, QLD",
    "200a Clarendn St, Sth Melbourne, VIC 3205"
]

for address in sample_data:
    parsed_address = parse_address(address, street_types)
    print(parsed_address)
```
